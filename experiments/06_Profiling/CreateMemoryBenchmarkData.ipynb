```python
from scanpy import read_h5ad
import anndata as ad
import pandas as pd
import os
import sys
```

```python
data_dir = "/home/julian/Uni/MasterThesis/data/"
reference_adata = read_h5ad(os.path.join(data_dir, "simdata_TS0.3_SS0.3.h5ad"))
```

```python
print(reference_adata.shape)
print(sys.getsizeof(reference_adata))
```

```python
n = 20

stacked_adata = reference_adata.copy()
for i in range(n):

    stacked_adata = ad.concat([stacked_adata, reference_adata])
    stacked_adata.obs_names_make_unique()

    print("Iteration %s, shape: %s, size (Mb): %s" % (i, stacked_adata.shape, sys.getsizeof(stacked_adata) / 1000000))

    outfile = os.path.join(data_dir, "mem_benchmark_%sx%s.h5ad" % (stacked_adata.shape[0], stacked_adata.shape[1]))
    stacked_adata.write_h5ad(outfile)
```
